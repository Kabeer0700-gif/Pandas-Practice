```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('anime.csv')
df.head()
```

```python
def extract_episode(txt) -> str:
    check = False
    data = ""
    for i in txt:
        if i == " ":
            if check:
                return data 
            
        if check:
            data += i
            
        if i == '(':
            check = True
```

```python
df['Episode']=df['Title'].apply(extract_episode)
df.head()
```

```python
df['Episode']=df['Episode'].astype(int)
type(df.loc[0]['Episode'])
```

```python
def extract_time(txt):
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]

            return data
        
```

```python
df['Timestamp']=df['Title'].apply(extract_time)
df.head()
```

```python
#anime with highest score
df[df['Score'] == df['Score'].max()]['Title']
```

```python
top_5 = df.sort_values(
    by="Score",
    ascending=False
).head(5)
top_5
```

```python
top_5_episode = df.sort_values(
    by="Episode",
    ascending=False
).head(5)

top_5_episode
```

```python
from datetime import datetime

def get_duration_months(period):
    start_str, end_str = period.split(" - ")

    start = datetime.strptime(start_str, "%b %Y")
    end = datetime.strptime(end_str, "%b %Y")

    return (
        (end.year - start.year) * 12
        + (end.month - start.month)
        + 1  # inclusive
    )

df["Duration_Months"] = df["Timestamp"].apply(get_duration_months)
df.head()
```

```python
df[df['Duration_Months'] == df['Duration_Months'].max()]['Title']
```

```python
top_5_running = df.sort_values(
    by="Duration_Months",
    ascending=False
).head(5)

top_5_running
```
